```python
import chardet

with open("most_streamed_2024.csv", "rb") as f:
    raw = f.read(200000)  # read a chunk
enc = chardet.detect(raw)["encoding"]
print("Detected:", enc)

df = pd.read_csv("most_streamed_2024.csv", encoding=enc)
print(df.columns)
df.head(3)
df.shape
```
